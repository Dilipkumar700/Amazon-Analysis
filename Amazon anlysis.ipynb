```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading data ##

```python
df = pd.read_csv('amazon.csv')
```

## Cleaning data ##

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['rating'].unique()
```

```python
df[df["rating"]=="|"]
```

```python
df[df['rating']!="|"]
```

```python
df['rating'] = pd.to_numeric(df['rating'],errors = "coerce")
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset = ['rating'])
```

```python
df.isnull().sum()
```

## Correcting Data Types ##

```python
df.dtypes
```

```python
df.head(2)
```

```python
df['discounted_price'] =pd.to_numeric(df['discounted_price'].str.replace(r"[^0-9]","",regex = True))
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace(r"[^0-9]","",regex = True))

df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace(r"[^0-9]","",regex = True))
df['rating_count'] =pd.to_numeric(df['rating_count'].str.replace(r"[^0-9]","",regex = True))
```

```python
df.dtypes
```

## Adding Columns ##

```python
df["user_count"] = df["user_id"].str.split(",").str.len()
```

```python
df.head(2)
```

## EDA Visualization ##

```python
plt.figure(figsize = (10,5))
sns.set_style("whitegrid")
```

```python
df['main_category'] = df['category'].str.split('|').str[0]

plt.figure(figsize=(10,6))
sns.countplot(y='main_category', data=df,
              order=df['main_category'].value_counts().index)
plt.title("Products by Main Category")
plt.show()
```

## Rating distribution ##

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title("Product Ratings Distribution")
plt.show()
```

## Discount percentange distribution ##

```python
plt.figure(figsize=(30,10))
sns.histplot(df['discount_percentage'], bins=20, kde=True)
plt.title("Discount Percentage")
plt.show()
```

## Actual price vs Discounted price ##

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='actual_price', y='discounted_price', data=df)
plt.title("Actual Price vs Discounted Price")
plt.show()
```

## Top 10 Reviewed Products ##

```python
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
top = df.nlargest(10,'rating_count')

plt.figure(figsize=(10,6))
sns.barplot(x='rating_count', y='product_name', data=top)
plt.title("Top Reviewed Products")
plt.show()
```

## Category vs Average rating ##

```python
avg = df.groupby('main_category')['rating'].mean().sort_values()

plt.figure(figsize=(10,6))
avg.plot(kind='barh')
plt.title("Average Rating by Category")
plt.show()
```

## Top 10 Expensive Products ##

```python
top_price = df.nlargest(10,'actual_price')

plt.figure(figsize=(10,6))
sns.barplot(x='actual_price', y='product_name', data=top_price)
plt.title("Most Expensive Products")
plt.show()
```

## Discount vs Rating ##

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='discount_percentage', y='rating', data=df)
plt.title("Discount vs Rating")
plt.show()
```

## Correaltion Heatmap ##

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

## Pie chart of Categories ##

```python
df['main_category'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title("Top Categories Share")
plt.ylabel('')
plt.show()
```
